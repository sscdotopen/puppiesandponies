```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data_dir = '/home/ssc/ownCloud/papers/realtime-recommendations/exploratory-analysis/'
figure_dir = '/home/ssc/ownCloud/papers/realtime-recommendations/figures/'

labelsize = 24
legendfontsize = 20
```

```python
degrees_1M = np.recfromcsv(data_dir  + 'movielens10M_itemdegrees_at1000000.tsv', delimiter='\t', names=['degree'])['degree']
degrees_5M = np.recfromcsv(data_dir  + 'movielens10M_itemdegrees_at5000000.tsv', delimiter='\t', names=['degree'])['degree']
degrees_10M = np.recfromcsv(data_dir  + 'movielens10M_itemdegrees_at10000000.tsv', delimiter='\t', names=['degree'])['degree']

cdf_1M, = plt.step(degrees_1M[::-1], np.arange(len(degrees_1M)) / float(len(degrees_1M)), linewidth='2') 
cdf_5M, = plt.step(degrees_5M[::-1], np.arange(len(degrees_5M)) / float(len(degrees_5M)), linewidth='2') 
cdf_10M, = plt.step(degrees_10M[::-1], np.arange(len(degrees_10M)) / float(len(degrees_10M)), linewidth='2') 


plt.yscale('log')
plt.xscale('log')
plt.xlabel('num interactions', fontsize=labelsize)
plt.ylabel('P(num interactions > x)', fontsize=labelsize)
plt.legend([cdf_1M, cdf_5M, cdf_10M], ['1M', '5M', '10M'], loc='lower left', fontsize=legendfontsize)
plt.savefig(figure_dir + 'movielens10M-degreedistributionovertime.pdf', dpi=300, format='pdf', bbox_inches='tight')
plt.show()
```
