```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data_dir = '/home/ssc/ownCloud/papers/realtime-recommendations/exploratory-analysis/'
figure_dir = '/home/ssc/ownCloud/papers/realtime-recommendations/figures/'

labelsize = 19
legendfontsize = 12
```

```python
#degrees_movielens1M = np.recfromcsv(data_dir  + 'movielens1M_itemdegrees.tsv', delimiter='\t', names=['degree'])['degree']
#degrees_movielens10M = np.recfromcsv(data_dir  + 'movielens10M_itemdegrees.tsv', delimiter='\t', names=['degree'])['degree']

degrees_lastfm = np.recfromcsv(data_dir  + 'lastfm_itemdegrees.tsv', delimiter='\t', names=['degree'])['degree']
degrees_stackoverflow = np.recfromcsv(data_dir  + 'stackoverflow_itemdegrees.tsv', delimiter='\t', names=['degree'])['degree']
degrees_dblp = np.recfromcsv(data_dir  + 'dblp_itemdegrees.tsv', delimiter='\t', names=['degree'])['degree']

#degrees_millionsongs = np.recfromcsv(data_dir  + 'millionsongs_itemdegrees.tsv', delimiter='\t', names=['degree'])['degree']
degrees_orkut = np.recfromcsv(data_dir  + 'orkut_itemdegrees.tsv', delimiter='\t', names=['degree'])['degree']
degrees_yahoomusic = np.recfromcsv(data_dir  + 'yahoomusic_itemdegrees.tsv', delimiter='\t', names=['degree'])['degree']

#cdf_movielens1M, = plt.step(degrees_movielens1M[::-1], np.arange(len(degrees_movielens1M)) / float(len(degrees_movielens1M)), linewidth='2') 
#cdf_movielens10M, = plt.step(degrees_movielens10M[::-1], np.arange(len(degrees_movielens10M)) / float(len(degrees_movielens10M)), linewidth='2', color='green') 
cdf_lastfm, = plt.step(degrees_lastfm[::-1], np.arange(len(degrees_lastfm)) / float(len(degrees_lastfm)), linewidth='2', color='red') 
cdf_stackoverflow, = plt.step(degrees_stackoverflow[::-1], np.arange(len(degrees_stackoverflow)) / float(len(degrees_stackoverflow)), linewidth='2', color='blue') 
cdf_dblp, = plt.step(degrees_dblp[::-1], np.arange(len(degrees_dblp)) / float(len(degrees_dblp)), linewidth='2', color='green') 
#cdf_millionsongs, = plt.step(degrees_millionsongs[::-1], np.arange(len(degrees_millionsongs)) / float(len(degrees_millionsongs)), linewidth='2') 
cdf_orkut, = plt.step(degrees_orkut[::-1], np.arange(len(degrees_orkut)) / float(len(degrees_orkut)), linewidth='2', color='black') 
cdf_yahoomusic, = plt.step(degrees_yahoomusic[::-1], np.arange(len(degrees_yahoomusic)) / float(len(degrees_yahoomusic)), linewidth='2', color='brown') 


plt.yscale('log')
plt.xscale('log')
plt.xlabel('num interactions', fontsize=labelsize)
plt.ylabel('P(num interactions > x)', fontsize=labelsize)
#plt.legend([cdf_movielens10M, cdf_lastfm, cdf_stackoverflow, cdf_orkut, cdf_yahoomusic], ['movielens10M', 'lastfm', 'stackoverflow', 'orkut', 'yahoomusic'], loc='lower left', fontsize=legendfontsize)
plt.legend([cdf_lastfm, cdf_stackoverflow, cdf_dblp, cdf_orkut, cdf_yahoomusic], ['lastfm', 'stackoverflow', 'dblp', 'orkut', 'yahoomusic'], loc='lower left', fontsize=legendfontsize)
plt.savefig(figure_dir + 'degreedistributions.pdf', dpi=300, format='pdf', bbox_inches='tight')
plt.show()
```
