```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data_dir = '/home/ssc/ownCloud/papers/realtime-recommendations/exploratory-analysis/'
```

```python
def load_stats(data_dir, file):
    return np.recfromcsv(data_dir + file, delimiter='\t', names=['num_interactions', 'mean', 'median', 'p90', 'p99', 'p999', 'max'])    
    
def plot_stats(name, stats, figure_name):
    labelfontsize = 24
    ticksize = 16
    #plt.title(name, fontsize=labelfontsize)
    mean, = plt.plot(stats['num_interactions'], stats['mean'], linewidth='2')
    median, = plt.plot(stats['num_interactions'], stats['median'], linewidth='2')
    p90, = plt.plot(stats['num_interactions'], stats['p90'], linewidth='2')
    p99, = plt.plot(stats['num_interactions'], stats['p99'], linewidth='2')
    p999, = plt.plot(stats['num_interactions'], stats['p999'], linewidth='2')
    max, = plt.plot(stats['num_interactions'], stats['max'], linewidth='2')

    #plt.yscale('log')
    plt.xlabel('# interactions observed', fontsize=labelfontsize)
    plt.ylabel('degree', fontsize=labelfontsize)
    plt.tick_params(axis='both', which='major', labelsize=ticksize)
    plt.tick_params(axis='both', which='minor', labelsize=ticksize)
    #plt.legend([mean, median, p90, p99, p999, max], ['mean', 'median', 'p90', 'p99', 'p99.9', 'max'], loc='upper left')
    plt.legend([max, p999, p99, p90, mean, median], ['max', 'p99.9', 'p99', 'p90', 'mean', 'median'], loc='upper left')
    
    ax = plt.gca()
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    
    plt.savefig('/home/ssc/ownCloud/papers/ted-sigmod/figures/' + figure_name +'-degreesovertime.pdf', dpi=300, format='pdf', bbox_inches='tight')
    plt.show()    
```

## Growth of Item Degrees

```python
plot_stats("movie degrees over time - Movielens1M", load_stats(data_dir, 'movielens1M_itemdegreesovertime_every10000.tsv'), "movielens1M-item")
```
