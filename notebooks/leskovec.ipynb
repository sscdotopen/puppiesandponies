```python
 # -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#plt.rc('font', family='Arial')

data_dir = '/home/ssc/ownCloud/papers/realtime-recommendations/exploratory-analysis/'
figure_dir = '/home/ssc/ownCloud/papers/realtime-recommendations/figures/'
labelsize = 19
annotationsize = 10
titlesize=24
```

```python
stackoverflow_growth = np.recfromcsv(data_dir  + 'stackoverflow_growth.tsv', delimiter='\t', names=['month', 'num_items', 'num_interactions'])
lastfm_growth = np.recfromcsv(data_dir  + 'lastfm_growth.tsv', delimiter='\t', names=['month', 'num_items', 'num_interactions'])
#movielens1M_growth = np.recfromcsv(data_dir  + 'movielens1M_growth.tsv', delimiter='\t', names=['month', 'num_items', 'num_interactions'])
#movielens10M_growth = np.recfromcsv(data_dir  + 'movielens10M_growth.tsv', delimiter='\t', names=['month', 'num_items', 'num_interactions'])
dblp_growth = np.recfromcsv(data_dir  + 'dblp_growth.tsv', delimiter='\t', names=['month', 'num_items', 'num_interactions'])
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(np.log(lastfm_growth['num_items']), np.log(lastfm_growth['num_interactions']))
print("LASTFM - slope:" + str(slope) + ", intercept: " + str(intercept) + ", R^2: " + str(r_value**2) + ', e^intercept: ' + str(np.exp(intercept)))
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(np.log(stackoverflow_growth['num_items']), np.log(stackoverflow_growth['num_interactions']))
print("STACKOVERFLOW - slope:" + str(slope) + ", intercept: " + str(intercept) + ", R^2: " + str(r_value**2) + ', e^intercept: ' + str(np.exp(intercept)))
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(np.log(dblp_growth['num_items']), np.log(dblp_growth['num_interactions']))
print("DBLP - slope:" + str(slope) + ", intercept: " + str(intercept) + ", R^2: " + str(r_value**2) + ', e^intercept: ' + str(np.exp(intercept)))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)



stackoverflow_growth_function = np.exp(0.0660141737527) * stackoverflow_growth['num_items'] ** 1.23221013364
p1, = ax1.plot(stackoverflow_growth['num_items'], stackoverflow_growth_function, linestyle='-', color='black')

p2 = ax1.scatter(stackoverflow_growth['num_items'], stackoverflow_growth['num_interactions'], color='red')

#2008-10-03 to 2011-11-30
ax1.text(stackoverflow_growth['num_items'][0], stackoverflow_growth['num_interactions'][0], 'Oct 2008', fontsize=annotationsize)
ax1.text(stackoverflow_growth['num_items'][-1], stackoverflow_growth['num_interactions'][-1], 'Nov 2011', fontsize=annotationsize)

ax1.set_title('stackoverflow', fontsize=titlesize)
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlabel('# posts', fontsize=labelsize)
ax1.set_ylabel('# favorites', fontsize=labelsize)
#ax1.set_ylim(-1, len(data))
ax1.set_xlim(0, 10**6)
    
ax1.legend([p2, p1], ['interactions', r'$= 1.068 \cdot x^{1.23} \,\, R^2 = 0.994$'], loc='lower right')


lastfm_growth_function = np.exp(-2.86338494701) * lastfm_growth['num_items'] ** 1.6302292506
p1, = ax2.plot(lastfm_growth['num_items'], lastfm_growth_function, linestyle='-', color='black')

p2 = ax2.scatter(lastfm_growth['num_items'], lastfm_growth['num_interactions'], color='red')

# lastfm: 2005-02-14 to 2013-09-29
ax2.text(lastfm_growth['num_items'][0], lastfm_growth['num_interactions'][0], 'Feb 2005', fontsize=annotationsize)
ax2.text(lastfm_growth['num_items'][-1], lastfm_growth['num_interactions'][-1], 'Sep 2013', fontsize=annotationsize)

ax2.set_title('lastFM', fontsize=titlesize)
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_ylabel('# listenings', fontsize=labelsize)
ax2.set_xlabel('# bands', fontsize=labelsize)
ax2.set_ylim(10**3, 10**8)

ax2.legend([p2, p1], ['interactions', r'$= 0.057 \cdot x^{1.63} \,\, R^2 = 0.997$'], loc='lower right')


dblp_growth_function = np.exp(-2.46362296658) * dblp_growth['num_items'] ** 1.35702959353
p1, = ax3.plot(dblp_growth['num_items'], dblp_growth_function, linestyle='-', color='black')

p2 = ax3.scatter(dblp_growth['num_items'], dblp_growth['num_interactions'], color='red')

# lastfm: 2005-02-14 to 2013-09-29
ax3.text(dblp_growth['num_items'][0], dblp_growth['num_interactions'][0], '1980', fontsize=annotationsize)
ax3.text(dblp_growth['num_items'][-1], dblp_growth['num_interactions'][-1], '2014', fontsize=annotationsize)

ax3.set_title('DBLP', fontsize=titlesize)
ax3.set_xscale('log')
ax3.set_yscale('log')
ax3.set_ylabel('# collaborations', fontsize=labelsize)
ax3.set_xlabel('# authors', fontsize=labelsize)
ax3.set_ylim(10**3, 10**8)

ax3.legend([p2, p1], ['interactions', r'$= 0.085 \cdot x^{1.36} \,\, R^2 = 0.999$'], loc='lower right')

fig.subplots_adjust(wspace=0.35, hspace=0.5)
fig.set_size_inches(18, 4, forward=True)

plt.savefig(figure_dir + 'powerlawgrowth.pdf', dpi=300, format='pdf', bbox_inches='tight')
plt.show()
```
