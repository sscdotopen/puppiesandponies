```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data_dir = '/home/ssc/ownCloud/papers/realtime-recommendations/exploratory-analysis/'
figure_dir = '/home/ssc/ownCloud/papers/realtime-recommendations/figures/'

labelsize = 16
legendfontsize = 10
```

```python
degrees_cutcounter_item100 = np.recfromcsv(data_dir  + 'cutcounter_itemdegrees100.tsv', delimiter='\t', names=['degree'])['degree']
degrees_cutcounter_item200 = np.recfromcsv(data_dir  + 'cutcounter_itemdegrees200.tsv', delimiter='\t', names=['degree'])['degree']
degrees_cutcounter_item300 = np.recfromcsv(data_dir  + 'cutcounter_itemdegrees300.tsv', delimiter='\t', names=['degree'])['degree']
degrees_cutcounter_item_all = np.recfromcsv(data_dir  + 'synth_itemdegrees.tsv', delimiter='\t', names=['degree'])['degree']
degrees_cutcounter_user100 = np.recfromcsv(data_dir  + 'cutcounter_userdegrees100.tsv', delimiter='\t', names=['degree'])['degree']
degrees_cutcounter_user200 = np.recfromcsv(data_dir  + 'cutcounter_userdegrees200.tsv', delimiter='\t', names=['degree'])['degree']
degrees_cutcounter_user300 = np.recfromcsv(data_dir  + 'cutcounter_userdegrees300.tsv', delimiter='\t', names=['degree'])['degree']
degrees_cutcounter_user_all = np.recfromcsv(data_dir  + 'synth_userdegrees.tsv', delimiter='\t', names=['degree'])['degree']

fig, (ax1, ax2) = plt.subplots(1, 2)

cdf_item100, = ax1.step(degrees_cutcounter_item100[::-1], np.arange(len(degrees_cutcounter_item100)) / float(len(degrees_cutcounter_item100)), linewidth='2', color='red') 
cdf_item200, = ax1.step(degrees_cutcounter_item200[::-1], np.arange(len(degrees_cutcounter_item200)) / float(len(degrees_cutcounter_item200)), linewidth='2', color='green') 
cdf_item300, = ax1.step(degrees_cutcounter_item300[::-1], np.arange(len(degrees_cutcounter_item300)) / float(len(degrees_cutcounter_item300)), linewidth='2', color='blue') 
cdf_itemall, = ax1.step(degrees_cutcounter_item_all[::-1], np.arange(len(degrees_cutcounter_item_all)) / float(len(degrees_cutcounter_item_all)), linewidth='2', color='black', linestyle='--') 
ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.set_xlabel('num interactions per item', fontsize=labelsize)
ax1.set_ylabel('P(num interactions > x)', fontsize=labelsize)
ax1.legend([cdf_item100, cdf_item200, cdf_item300, cdf_itemall], [r'$k_{max}, f_{max} = 100$', r'$k_{max}, f_{max} = 200$', r'$k_{max}, f_{max} = 300$', 'unsampled'], loc='lower left', fontsize=legendfontsize)
ax1.set_xlim(0, 2 * 10**5)

cdf_user100, = ax2.step(degrees_cutcounter_user100[::-1], np.arange(len(degrees_cutcounter_user100)) / float(len(degrees_cutcounter_user100)), linewidth='2', color='red') 
cdf_user200, = ax2.step(degrees_cutcounter_user200[::-1], np.arange(len(degrees_cutcounter_user200)) / float(len(degrees_cutcounter_user200)), linewidth='2', color='green') 
cdf_user300, = ax2.step(degrees_cutcounter_user300[::-1], np.arange(len(degrees_cutcounter_user300)) / float(len(degrees_cutcounter_user300)), linewidth='2', color='blue') 
cdf_userall, = ax2.step(degrees_cutcounter_user_all[::-1], np.arange(len(degrees_cutcounter_user_all)) / float(len(degrees_cutcounter_user_all)), linewidth='2', color='black', linestyle='--') 
ax2.set_yscale('log')
ax2.set_xscale('log')
ax2.set_xlabel('num interactions per user', fontsize=labelsize)
ax2.set_ylabel('P(num interactions > x)', fontsize=labelsize)
ax2.legend([cdf_user100, cdf_user200, cdf_user300, cdf_userall], [r'$k_{max}, f_{max} = 100$', r'$k_{max}, f_{max} = 200$', r'$k_{max}, f_{max} = 300$', 'unsampled'], loc='lower left', fontsize=legendfontsize)

fig.subplots_adjust(wspace=0.25, hspace=0.5)
fig.set_size_inches(12, 3, forward=True)

plt.savefig(figure_dir + 'cutcounter-degreedistributions.pdf', dpi=300, format='pdf', bbox_inches='tight')
plt.show()
```
